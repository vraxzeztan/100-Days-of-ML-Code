```python
data = pd.read_csv("twitter_train.csv")
test = pd.read_csv("twitter_test.csv")
```

```python
y = data.airline_sentiment
```

```python
y_train = np.array(y)
y_train
```

```python
import nltk
nltk.download("popular")
```

```python
a = []
for i in range(len(data.text)):
    a.append((word_tokenize(data.text[i]),y[i]))
```

```python
a[0][1]
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
import string
punctuations = list(string.punctuation)
stop = stop + punctuations
num = ['1','2','3','4','5','6','7','8','9']
stop += num
```

```python
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
lemmatizer = WordNetLemmatizer()
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stop:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
a = [(clean_review(document),category) for document,category in a]
```

```python
aa = []
for i in range(len(test.text)):
    aa.append(word_tokenize(test.text[i]))

aa = [clean_review(document) for document in aa]
```

```python
x_train = [" ".join(document) for document, category in a]
x_test = [" ".join(document) for document in aa]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vec = CountVectorizer(max_features = 2000, ngram_range=(1,3))
x_train_features = count_vec.fit_transform(x_train)
x_train_features.todense()
```

```python
count_vec.get_feature_names()
```

```python
x_test_features = count_vec.transform(x_test)
```

```python
from sklearn.svm import SVC
svc = SVC(C= 100, gamma=0.0008)
svc.fit(x_train_features, y_train)
```

```python
svc.score(x_train_features, y_train)
```

```python
y_test = svc.predict(x_test_features)
```

```python
np.savetxt('twitter_pred.csv',y_test,fmt='%s')
```
